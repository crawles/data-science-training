## Step 2
Prepare data to 1D array

```python
#TODO: trap dummy variable
# we add ID in bc, later, we will put this in a pandas dataframe in plpython
# and return (id, score)
features = ['id', 'longitude','latitude', 'year', 'month', 'day', 'day_of_year',\
 'species_CULEX ERRATICUS', 'species_CULEX PIPIENS', 'species_CULEX PIPIENS/RESTUANS',\
 'species_CULEX RESTUANS', 'species_CULEX SALINARIUS', 'species_CULEX TARSALIS',\
 'species_CULEX TERRITANS', 'avg', 'preciptotal7', 'tmin7', 'tavg7', 'tmax7']
feature_formatted = ['"{}"'.format(v) for v in features]
feature_str = ','.join(feature_formatted)
```

```python
%%read_sql
DROP AGGREGATE IF EXISTS array_agg_array(anyarray);
CREATE AGGREGATE array_agg_array(anyarray)
(
    SFUNC = array_cat,
    STYPE = anyarray
);
```

### Create training data

```python
output_table = 'wnv.training_array'
```

```python
%%read_sql
DROP TABLE IF EXISTS {output_table};
CREATE TABLE {output_table}
AS
WITH training_table_arrayed AS (SELECT array[{feature_str}] a,
                                      wnv_present
                               FROM {feature_table_tt}
                               WHERE testing_example = 0)
SELECT array_agg_array(a) features, array_agg(wnv_present) wnv_present
FROM training_table_arrayed;
```

```python
output_table = 'wnv.testing_array'
num_bins = 10
```

```python
%%read_sql
DROP TABLE IF EXISTS {output_table};
CREATE TABLE {output_table}
AS
WITH training_table_arrayed AS (SELECT row_number() over ()%{num_bins} AS bin_id,
                                       array[{feature_str}] a,
                                       wnv_present
                                FROM {feature_table_tt}
                                WHERE testing_example = 0)
SELECT bin_id, array_agg_array(a) features, array_agg(wnv_present) wnv_present
FROM training_table_arrayed
GROUP BY bin_id;
```

```python
%%read_sql
-- return type for modeling
DROP TYPE IF EXISTS result_row CASCADE;
CREATE TYPE result_row AS (
    id text,
    score float8
);
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
cl = RandomForestClassifier()
X = df[df.columns[10:]]
X = df.drop(['date_ts', 'wnv_present', 'species', 'trap', 'total_num_mosquitos'], axis = 1)
y = df.wnv_present
X_train,X_test,y_train,y_test = train_test_split(X,y)
cl.fit(X_train, y_train)
```

```python
y_prob = cl.predict_proba(X_test)[:,1]
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_prob)
```

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
X = X_train
importances = cl.feature_importances_
std = np.std([tree.feature_importances_ for tree in cl.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.barh(range(X.shape[1])[::-1], importances[indices],
       color="r", yerr=std[indices], align="center")
plt.yticks(range(X.shape[1])[::-1], X.columns[indices])
plt.ylim([X.shape[1] - 25, X.shape[1]])
plt.show()
```
